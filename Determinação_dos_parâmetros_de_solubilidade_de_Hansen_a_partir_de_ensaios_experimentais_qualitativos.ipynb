##Teoria da solubilidade de Hildebrand

A energia relacionada a solubilidade de uma substância em um solvente é a necessária para que a interação entre as moléculas dessa substância entre si seja, no mínimo, a mesma entre elas e as moléculas do solvente. Tomando a entalpia de vaporização $h_v\ $como a energia limite de interação entre as moléculas de uma substância e levando em consideração a forma e o tamanho da molécula como elementos importantes na sua mobilidade e interação na mistura, Hildebrand propôs que a energia de solubilização $\mathrm{E}_s\ $ (1) pra essa substância seja a razão entre a sua entalpia de vaporização e o volume molar da substância, como em (2), criando o parâmetro de solubilidade $\delta$.

\begin{align}
    \\
    \mathrm{E}_s = h_v &&(1)\\
    \\
    \delta^2 = \frac{\mathrm{E}_s}{V_m} = \frac{h_v}{V_m} &&(2)\\
    \\
    \end{align}
    
Grandeza | símbolo
--- | :---: 
`Energia de solubilização` | $\mathrm{E}_s$ 
`Volume molar` | $V_m$ 
`Entalpia de vaporização` | $h_v$
`Parâmetro de Solubilidade` | $\delta$
$\ \$

Moléculas com um parâmetro de solubilidade $\delta\ $ semelhante possuiriam energias de solubilização semelhantes e, por isso, teriam facilidade em interagir, situação na qual a mistura formaria uma solução. Quanto mais próximo o $\delta\ $ entre elas, maior a facilidade de formação de solução, sendo o oposto também verdadeiro: quanto mais distantes os parâmetros de solubilidade do solvente e do soluto, menos compatíveis as substâncias e menor a probabilidade de formação de solução entre elas.

## Parâmetros de solubilidade de Hansen
Hansen refinou a proposta original de Hildebrand, propondo que o parâmetro de solubilidade total $\delta_t\ $ seria composto pela soma das contribuições das energias (3) de dispersão, polares e de interações de hidrogênio (4) entre as moléculas, como em (5).

\begin{align}
    \\
    \mathrm{E}_s = h_v = \mathrm{E}_d + \mathrm{E}_p + \mathrm{E}_h &&(3)\\
    \\
    \delta_d^2=\frac{\mathrm{E}_d}{V_m}\ \text{, }\ \ 
    \delta_p^2=\frac{\mathrm{E}_p}{V_m}\ \ \text{ e }\ \ 
    \delta_h^2=\frac{\mathrm{E}_h}{V_m}\ \ \text{, } &&(4)\\
    \\
    \delta^2_t = \delta_d^2 + \delta_p^2 + \delta_h^2 &&(5)\\
    \\
    \end{align}

Grandeza | símbolo
--- | :---: 
`Energia de solubilização` | $\mathrm{E}_s$ 
`Parcela de Energia de forças de dispersão` | $\mathrm{E}_d$ 
`Parcela de Energia de forças de polares` | $\mathrm{E}_p$
`Parcela de Energia de forças de interação de hidrogênios` | $\mathrm{E}_h$
`Volume molar` | $V_m$ 
`Entalpia de vaporização` | $h_v$
`Parâmetro de Solubilidade` | $\delta$
`Parâmetro de Solubilidade total` | $\delta_t$
`Contribuição das forças de dispersão` | $\delta_d$
`Contribuição das forças polares` | $\delta_p$
`Contribuição das forças de interação de hidrogênios` | $\delta_h$
$\ \$

Tomando a distância entre os parâmetros de solubilidade total $\mathrm{R}_a\ $ de duas substâncias como uma diferença vetorial (6), e estabelecendo uma região próxima do soluto como propícia para a interação, Hansen traz a "esfera de solubilidade" $\mathrm{R}_o\ $, volume no qual toda substância contida interagiria forte o suficiente com o soluto, formando uma solução. Sendo a distância vetorial entre os parâmetros de solubilidade total das substâncias sua diferença de energia e considerando o raio da esfera de solubilidade do soluto como referência, pode-se calcular a diferença relativa de energia $\mathrm{RED}$ (7) entre o soluto e o solvente. Na mistura, quando $\mathrm{RED} \le 1\ $  há a formação da solução.


\begin{align}
        \\
        \mathrm{R}_a = \sqrt{(\delta_{d_{soluto}}-\delta_{d_{solvente}})^2+(\delta_{p_{soluto}}-\delta_{p_{solvente}})^2+(\delta_{h_{soluto}}-\delta_{h_{solvente}})^2}   &&(6) \\
        \\
        \mathrm{RED} = \frac{\mathrm{R}_a}{\mathrm{R}_o} && (7) \\
        \\
        \end{align}


A determinação do valor do raio da esfera de solubilidade é feita à partir de ensaios experimentais de solubilidade, onde se adota para $\mathrm{R}_o\ $ o valor da maior distância $\mathrm{R}_a\ $ entre os solventes capazes de dissolver o soluto $\mathrm{RED}\le 1$.

##Calcular tíner não é tão complicado com o HSP.
Os compostos dos solventes (tíneres) foram calculados geométricamente em relação à posição de cada componente da mistura entre os solventes puros, conforme a proporção de concentração no novo tíner, obtendo-se um novo conjunto de parâmetros de solubilidade para cada tíner.

HSP of Solvents Mixture : https://www.pirika.com/ENG/HSP/JAVA/GSD/GSD.html visitado em 20/01/22.

\begin{align}\\
        (\delta_{d_m},\ \delta_{p_m},\ \delta_{h_m}) = \frac{(a\delta_{d_1}+b\delta_{d_2})}{a+b}, \frac{(a\delta_{p_1}+b\delta_{p_2})}{a+b}, \frac{(a\delta_{h_1}+b\delta_{h_2})}{a+b} &&(8)
      \end{align}

Então foi procurada na tabela **biblio_semaforo_merged** os reagentes para os tíneres e calculados os novos compostos solventes.

````python
#@title
# Os compostos, que precisam ser construídos à partir da mistura entre os solventes puros são:
  # Ethanol 96%
  # Ethanol 99,9%
  # Ácido Fórmico 90%

# Localiza os reagentes de partida para o cálculo dos tíneres:
print(biblio_solventesAbbott[(biblio_solventesAbbott.index.isin([1211, 1209, 645, 734]))])#[['Nome', 'dd', 'dp', 'dh']])
````

````python
#@title
# Atribui os valores de dd, dp e dh de cada reagente de partida a uma variável correspondente:
water =  biblio_solventesAbbott[(biblio_solventesAbbott.index.isin([1209]))][['dd', 'dp', 'dh', 'Smile','CID','CAS','Apelido','classe','Metodo', 'Nome']] # 1211 Water Complete Misc. (R=13.0)
waterH = biblio_solventesAbbott[(biblio_solventesAbbott.index.isin([1211]))][['dd', 'dp', 'dh', 'Smile','CID','CAS','Apelido','classe','Metodo', 'Nome']] # 1209 Water
ethanol = biblio_solventesAbbott[(biblio_solventesAbbott.index.isin([645]))][['dd', 'dp', 'dh', 'Smile','CID','CAS','Apelido','classe','Metodo', 'Nome']] # 645 Ethanol
formic_acid = biblio_solventesAbbott[(biblio_solventesAbbott.index.isin([734]))][['dd', 'dp', 'dh', 'Smile','CID','CAS','Apelido','classe','Metodo', 'Nome']] # 734 Formic Acid

# Função que calcula o tíner:
def thiner(solvA, solvB, ConcA):
  '''
    thiner([solvA_dd, solvA_dp,solvA_dh],[solvB_dd, solvB_dp,solvB_dh], Concentração_de_A_em_B )
      
    ref: HSP of Solvents Mixture : https://www.pirika.com/ENG/HSP/JAVA/GSD/GSD.html visitado em 20/01/22
  '''
  dda, dpa, dha, sa, ca, caa, apa, cla, mea, na = solvA
  ddb, dpb, dhb, sb, cb, cab, apb, clb, meb, nb = solvB
  na = na+' '+str(ConcA)+'%'
  a = float(ConcA)
  b = 100.0-a
  ddm, dpm, dhm = [(a*ddb+b*dda)/(a+b), (a*dpb+b*dpa)/(a+b),(a*dhb+b*dha)/(a+b)] 
  return [ddm, dpm, dhm, sa, ca, caa, apa, cla, mea, na] 

print( 'HSP valores dos solventes puros: ')
print('HSP Ethanol: ', ethanol.values[0])
print('HSP Formic Acid: ', formic_acid.values[0])
print('HSP Water:', water.values[0])
print('HSP Water Comp. Misc.:', waterH.values[0])
print('')
Ethanol_96_w = thiner(ethanol.values[0], water.values[0], 96.0)
Ethanol_999_w = thiner(ethanol.values[0], water.values[0], 99.9)
Formic_Acid_90_w = thiner(formic_acid.values[0], water.values[0], 90.0)
Ethanol_96_wh = thiner(ethanol.values[0], waterH.values[0], 96.0)
Ethanol_999_wh = thiner(ethanol.values[0], waterH.values[0], 99.9)
Formic_Acid_90_wh = thiner(formic_acid.values[0], waterH.values[0], 90.0)
print('HSP para os tíneres:')
print('Ethanol_96_w ', Ethanol_96_w)
print('Ethanol_999_w ', Ethanol_999_w)
print('Formic_Acid_90_w ', Formic_Acid_90_w)
print('Ethanol_96_wh', Ethanol_96_wh)
print('Ethanol_999_wh', Ethanol_999_wh)
print('Formic_Acid_90_wh', Formic_Acid_90_wh)
print('')
````

##Agora é otimizar!
Com os dados experimentais publicados por Hansen em seu Handbook e os valores dos parâmetros HSP obtidos por ele, chegou a hora de tentar reproduzir a função "desirability function",  buscando implementar o método de otimização estatística proposto no Handbook.

A idéia é ter uma rotina capaz de obter os mesmos resultados, ou resultados próximos e com menos erros em relação ao comportamento experimental relatado originalmente, à partir dos mesmos dados experimentais.

##'*Desirability Function*'
Hansen propõe que os parâmetros de solubilidade de uma substância podem ser obtidos através de ensaios qualitativos de solubilidade e selecionou um grupo de 88 solventes cujos parâmetros de solubilidade já são determinados, nos quais ele testou um grupo de 33 polímeros.

A cada um dos polímeros e em cada um dos solventes, Hansen atribuiu um valor qualitativo de 1 a 6 que descrevia o comportamento do polímero no solvente, de acordo com a tabela:

```
(obtido no Handbook do Hansen, pág. 507, Apendix A, Table A3)
1. Soluble
2. Almost soluble
3. Strongly swollen, slight solubility
4. Swollen
5. Little swelling
6. No visible effect
```
Pode ser traduzido como
```
1. Solúvel
2. Quase solúvel
3. Fortemente inchado, leve solubilidade
4. Inchado
5. Pequeno inchaço
6. Nenhum efeito visível
```
Todo o conjunto de dados experimentais publicados por Hansen em seu Handbook estão contidos no dataframe **semaforo_biblio_merged**. É esse dataframe que será utilizado como entrada na função de otimização.

Para cada polímero, será feita uma chamada à função de otimização, para a qual é fornecida uma lista dos solventes testados com aquele polímero e o valor atribuído no semáforo ao comportamento identificado na mistura. Também é fornecida à função de otimização um valor inicial para os parâmetros HSP e o raio da esfera de solubilidade para este polímero, que deverá ser otimizado.

Estes valores iniciais são muito importantes, pois influenciam o resultado da otimização estatística. No entanto, Hansen não deixa muito claro no Handbook os critérios e quais valores iniciais ele utilizou.

Aqui vamos utilizar a mediana dos parâmetros dos solventes marcados como "1. soluble" para realizar a primeira otimização.

Com base nesses valores iniciais arbitrários, a função de objetivo calcula a distância ($\mathrm{R}_a$) e a ($\mathrm{RED}$) diferença relativa de energia entre o polímero e cada solvente do conjunto, simulando para este valor inicial a disposição do espaço de Hansen do polímero. A maior distância entre o soluto e o solvente marcado como bom solvente no semáforo é adotada como raio ($\mathrm{R}_o$) da esfera de solubilidade de Hansen para este soluto. Nesta simulação também são computados os erros, solventes que estão marcados no semáforo como bons solventes posicionados fora da esfera de solubilidade do polímero e também o contrário, solventes marcados como incapazes de dissolver o polímero posicionados dentro do semáforo. O retorno desta função de otimização é um valor escalar entre 0 e 1, sendo zero o pior conjunto de parâmetros ($\delta_d$, $\delta_p$, $\delta_h$, $\mathrm{R}_o$) em relação aos dados experimentais e 1 o melhor conjunto de parâmetros para representar o observado no experimento de solubilidade.

A biblioteca de otimização scipy.minimize() é a encarregada de iterar parâmetros na função objetivo, buscando o conjunto de parâmetros onde a função objetivo tem seu máximo e para isso conta com vários algoritmos matemáticos que podem ser utilizados. As bibliotecas do Python utilizadas neste trabalho permitem a execução dos algoritmos "Simplex" (nelder-mead), "COBYLA", "SLSQP" e "BFGS" neste processo. Verificamos que o algoritmo "COBYLA" apresentou a melhor relação custo (tempo de execução) x benefício e por isso foi escolhida como padrão. 

O retorno da biblioteca de otimização é o conjunto de parâmetros HSP procurado para o soluto.

Comparando o resultado desse algoritmo com os resultados publicados por Hansen, queremos estabelecer sua confiança e acuracidade na interpretação dos dados experimentais deste estudo.

Aqui, organizamos a entrada dos dados experimentais na função:

Adaptando a função de desejabilidade original (9) (10), Hansen propõe a análise dos dados experimentais de solubilidade tabulados como um semáforo que varia entre 1 a 6, sendo 1 totalmente solúvel e 6 totalmente insolúvel na forma de (11) e (12) procurando $\mathrm{D}\ $ o mais próximo possível de $1$.

*Desirability Function* original:

\begin{align}\\
        {\mathrm{A}_i(\hat{y}_i(x))} = \left\{
        \begin{array}{ll}
        0 & && \text{Se}\ \hat{y_i}(x)\ \le Y^{min}_i\ \ \text{ou}\ \ \hat{y_i}(x)> Y^{máx}_i, \\
        \left[ \frac{\hat{y_i}{}(x)-Y^{min}_i}{T^{min}_i - Y^{min}_i} \right]^{s_i} & && \text{Se}\ Y^{min}_i\ < \hat{y_i}(x) \le T^{min}_i,\\
        \left[ \frac{Y^{máx}_i - \hat{y_i}{}(x)}{Y^{máx}_i - T^{máx}_i} \right]^{t_i} & && \text{Se}\ T^{máx}_i\ \le \hat{y_i}(x) \le Y^{máx}_i,\\
        1 & && \text{Se}\ T^{min}_i\ < \hat{y_i}(x) \le T^{máx}_i.\\[1em]
        \end{array}
        \right. &&(9)\\[2em]
        \\
        \end{align}

\begin{align}
        \text{D}\ = \left\{{\prod^n_{i\in D}}(\mathrm{A_i}^{w_i}) ^{\frac{1}{\sum_{i=1}^i{w_i}}}
         \right. &&(10)\\
         \\ 
        \text{Onde }0 < \mathrm{A}_i < 1 \text{ e }0 < \mathrm{D} < 1.\\
        \\
    \end{align}



*Desirability Function* modificada para obter os parâmetros de Hansen a partir do semáforo gerado em ensaio experimental qualitativo de solubilidade:

\begin{align}\\
\\
        {\mathrm{A}_i} = \left\{
        \begin{array}{ll}
        1 &&\text{Se }\mathrm{RED} <=\ 1\text{ e Semáforo <=2 ou}\\
        && \text{Se } \mathrm{RED} >\ 1\text{ e Semáforo >2.} \\
        \mathrm{e}_i^{-|[(\mathrm{R}_a-\mathrm{R}_o)^\frac{1}{w_i}]| ^ { 0.58*\mathrm{e}^{erros^{R_o}} }  }\\
        \end{array}
        \\
        \right. &&(11)\\[2em]
        \end{align}

\begin{align}
        \text{D}\ = \left\{
          {\prod^n_{i\in D}} \mathrm{A_i}^{\frac{1}{\sum_{i=1}^i{w_i}}}
         \right. &&(12)\\
         \\ 
    \end{align}
                                     )

````python
# Compara os valores e calcula o erro na tabela dfol.
# Para cada linha em dfol com método COBYLA, conta os erros em relação a lista semáforo experimental da bibliografia.
# Faz o mesmo para cada linha em dfol com método Hansen, que é o original da bibliografia
````

````python
def compara_erros(soluto, otimizado, biblio, min_solub):
  bd=[]
  teste = soluto
  dfol = otimizado
  dfu_A = biblio
  min_solub = min_solub
  print('Polímero:', teste)
  dd = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['COBYLA']))].dd.values[0]
  dp = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['COBYLA']))].dp.values[0]
  dh = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['COBYLA']))].dh.values[0]
  Ro = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['COBYLA']))].Ro.values[0]
  print('Otimizado aqui: ',dd, dp, dh, Ro)

  ddB = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['Hansen']))].dd.values[0]
  dpB = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['Hansen']))].dp.values[0]
  dhB = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['Hansen']))].dh.values[0]
  RoB = dfol[(dfol['Nome'].isin([teste]) & dfol['Metodo'].isin(['Hansen']))].Ro.values[0]
  print('Parâmetros da bibliografia: ', ddB, dpB, dhB, RoB)

  lista = dfu_A[(~dfu_A[teste].isna())][[teste,'dd','dp','dh']]
  #print(lista.shape)
  #print(lista)
  lista['Ra_HSPmol'] = ((4.0*((lista.dd - dd)**2.0)) + ((lista.dp - dp)**2.0) + ((lista.dh - dh)**2.0))**(1/2)
  lista['RED_HSPmol'] = lista['Ra_HSPmol']/Ro
  lista['erro_HSPmol'] = np.where(((lista['RED_HSPmol'] <= 1) & (lista[teste] < min_solub)) | ((lista['RED_HSPmol'] > 1) & (lista[teste] > min_solub)), 0, 1 )

  lista['Ra_Hansen'] = ((4.0*((lista.dd - ddB)**2.0)) + ((lista.dp - dpB)**2.0) + ((lista.dh - dhB)**2.0))**(1/2)
  lista['RED_Hansen'] = lista['Ra_Hansen']/RoB
  #print(lista)
  lista['erro_Hansen'] = np.where(((lista['RED_Hansen'] <= 1) & (lista[teste] < min_solub)) | ((lista['RED_Hansen'] > 1) & (lista[teste] > min_solub)), 0, 1 )

  if lista['erro_HSPmol'].sum()>0:
    diferenca = ((lista['erro_HSPmol'].sum() - lista['erro_Hansen'].sum() )/lista['erro_HSPmol'].sum() )*100
  else:
    diferenca = 0

  MAPE_Hansen = MAPE(lista[(lista[teste]<min_solub)].shape[0], lista[(lista['RED_Hansen'] <= 1)].shape[0]) # Calcula a acurácia do modelo (MAPE % - erro percentual absoluto médio)
  MAPE_HSPmol = MAPE(lista[(lista[teste]<min_solub)].shape[0], lista[(lista['RED_HSPmol'] <= 1)].shape[0]) # Calcula a acurácia do modelo (MAPE % - erro percentual absoluto médio)
  total_erros_otimizado = lista['erro_HSPmol'].sum()
  total_erros_bibliografia = lista['erro_Hansen'].sum()

  #print(lista)
  print('MAPE Hansen: ', MAPE_Hansen, ' MAPE HSPmol: ', MAPE_HSPmol) # Calcula a acurácia do modelo (MAPE % - erro percentual absoluto médio)
  print('Polímero: ', teste, ', Total de solventes no ensaio: ',lista.shape[0],', Total de erros no otimizado: ',total_erros_otimizado, ', Total de erros na bibliografia: ',total_erros_bibliografia, ', Diferença: ', diferenca, '%')
  bd = [{'polimero':teste, 'total_erros_otimizado':total_erros_otimizado, 'MAPE_HSPmol':MAPE_HSPmol,
         'total_erros_bibliografia':total_erros_bibliografia, 'MAPE_Hansen':MAPE_Hansen, 'diferenca_percentual':diferenca }]
  # print(bd) # Lista que será inserida no dataframe de referência.
  return diferenca, lista, bd

df_compara = pd.DataFrame(columns=['polimero', 'total_erros_otimizado', 'MAPE_HSPmol', 'total_erros_bibliografia', 'MAPE_Hansen', 'diferenca_percentual'])
for s in semaforos:#['A']:#
  print('')
  diferenca, lista, bd = compara_erros(s, dfol, semaforo_biblio_merged, min_solub)
  #df_compara = df_compara.append(bd, ignore_index=True)
  df_compara = df_compara.append(bd, ignore_index=True)
````
